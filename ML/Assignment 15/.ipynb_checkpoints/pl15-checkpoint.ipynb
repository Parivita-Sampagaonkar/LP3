Title: Implementation of Anomaly Detection

```python
# ================================================================
#  Practical 15
# Title: Implementation of Anomaly Detection using Local Outlier Factor
# ================================================================

# ------------------------------------------------
# 1. Import Required Libraries
# ------------------------------------------------
from sklearn.neighbors import LocalOutlierFactor   # For outlier detection
from sklearn.datasets import make_blobs            # For generating synthetic data
from numpy import quantile, where                  # For quantile thresholding
import matplotlib.pyplot as plt                    # For visualization
import numpy as np                                 # For numerical operations

# ------------------------------------------------
# 2. Prepare Synthetic Dataset
# ------------------------------------------------
np.random.seed(1)  # For reproducibility

# Generate 1000 samples around one center (normal points)
X, _ = make_blobs(
    n_samples=1000, centers=1, cluster_std=0.3, center_box=(0, 10), random_state=42
)

# Visualize dataset
plt.scatter(X[:, 0], X[:, 1], edgecolor='k', s=15)
plt.title("Generated Data Points")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# ------------------------------------------------
# 3. Apply Local Outlier Factor (LOF)
# ------------------------------------------------
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.03)
y_pred = lof.fit_predict(X)

# Identify the indices of the outliers in the dataset
outlier_indices = where(y_pred == -1)
outlier_values = X[outlier_indices]

# Visualize detected outliers
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], label='Normal Points', edgecolor='k', s=15, alpha=0.7)
plt.scatter(outlier_values[:, 0], outlier_values[:, 1], color='r',
            label='Outliers', edgecolor='k', s=30)
plt.title("Outlier Detection using Local Outlier Factor (LOF)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(loc="best")
plt.show()

# ------------------------------------------------
# 4. Anomaly Detection using LOF Scores
# ------------------------------------------------
model = LocalOutlierFactor(n_neighbors=20)
model.fit_predict(X)

# Access the negative outlier factor (the lower, the more abnormal)
lof_scores = model.negative_outlier_factor_

# Define threshold at 3rd percentile
threshold = quantile(lof_scores, 0.03)
print(f"Calculated LOF Threshold: {threshold:.4f}")

# Identify points below threshold as outliers
index = where(lof_scores <= threshold)
values = X[index]

# ------------------------------------------------
# 5. Visualization of Outliers
# ------------------------------------------------
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], label='Normal Points', edgecolor='k', s=15, alpha=0.7)
plt.scatter(values[:, 0], values[:, 1], color='red', label='Detected Outliers',
            edgecolor='k', s=30)
plt.title('Scatter Plot of Data Points with Outliers Highlighted')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend(loc='best')
plt.show()

print(f"\n Total Outliers Detected: {len(values)}")
print(" Practical Completed Successfully!")
```
