```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = np.array([
    [5, 3, 0, 1],
    [4, 0, 0, 1],
    [1, 1, 0, 5],
    [1, 0, 0, 4],
    [0, 1, 5, 4],
], dtype=float)
```

```python
print("ðŸŽ¬ Original Rating Matrix (Missing values as 0):")
print(R)
```

```python
# --- Step 3: Apply Singular Value Decomposition (SVD) ---
U, sigma, Vt = np.linalg.svd(R, full_matrices=False)
```

```python
print("\nðŸ§© User Feature Matrix (U):")
print(np.round(U, 3))

print("\nðŸ”¹ Singular Values (Î£):")
print(np.round(sigma, 3))

print("\nðŸ§© Item Feature Matrix (Vt):")
print(np.round(Vt, 3))
```

```python
# --- Step 4: Choose Reduced Dimension (Rank-k) ---
k = 2  # can adjust this value for different approximations
sigma_k = np.diag(sigma[:k])
```

```python
# --- Step 5: Approximate Original Matrix using Top-k Features ---
R_approx = np.dot(U[:, :k], np.dot(sigma_k, Vt[:k, :]))
```

```python
print("\nðŸ”® Approximated Rating Matrix (Rank-{}):".format(k))
print(np.round(R_approx, 2))
```

```python
# --- Step 6: Replace Missing (0) Entries with Predicted Values ---
R_pred = R.copy()
R_pred[R_pred == 0] = R_approx[R_pred == 0]
R_pred = np.clip(R_pred, 1, 5)  # keep ratings in 1â€“5 range

print("\nðŸ“Š Final Predicted Rating Matrix:")
print(np.round(R_pred, 2))
```

```python
# --- Step 7: Visualize Original vs Approximated Matrices ---
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(R, cmap="Blues")
axes[0].set_title("Original Rating Matrix")
axes[0].axis("off")

axes[1].imshow(R_approx, cmap="Blues")
axes[1].set_title(f"Approximated Matrix (Rank-{k})")
axes[1].axis("off")

plt.show()
```
